**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
a = np.random.randint(1, 10, size = 10) 
print(a)

mask = (a > 3) & (a < 8)

a[mask] = -a[mask]
        
print("result:\n", a)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
a = np.random.randint(1, 10, size = 10) 

print(a)

a[np.argmax(a)] = 0

print("result:\n", a)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
a = np.array(np.random.randint(1, 10, size = (2, 2)))
print(a)

n_rows, n_cols = a.shape
# все комбинации индексов
indices = np.indices((n_rows,) * n_cols)
indices = indices.reshape(n_cols, -1).T

result = np.zeros((n_rows ** n_cols, n_cols), dtype=int)
for i, idx_combo in enumerate(indices):
    result[i] = a[idx_combo, range(n_cols)]

print("result\n", result)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.array([
    [9, 1, 4],
    [1, 3, 4],
    [7, 9, 6],
    [8, 4, 3],
    [4, 6, 6],
    [6, 3, 9],
    [2, 7, 2],
    [5, 6, 9]
])
B = np.array([
    [3, 4],
    [1, 9]
])

masks = []
for row_b in B:
    # проверка для каждого элемента B есть ли он в A
    element_masks = [np.any(A == element, axis=1) for element in row_b]
    
    # строка A должна содержать хотя бы один элемент из текущей строки B
    masks.append(np.any(element_masks, axis=0))

final_mask = np.all(masks, axis=0)
result = A[final_mask]

print('matching rows:')
print(result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# without numpy
X = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]

print(X)

rows = len(X)
cols = len(X[0])

size_diag = min(rows, cols)

result = 1

# проходим по диагональным элементам
for i in range(size_diag):
    if X[i][i] != 0:
        result *= X[i][i]
    
print("result: ", result)
```

```python
# numpy

X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])

diagonal = X.diagonal()

result = np.prod(diagonal[diagonal != 0])

print(result)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
## without numpy

x = [2, 2, 2, 3, 3, 3, 5, 2, 1]

nums, count = [], []

if x:
    current_num = x[0]   # начинаем с первого элемента
    current_count = 1

    for i in range(1, len(x)):
        if x[i] == current_num:
            current_count += 1  # увеличиваем счетчик для текущего числа
        else:
            # сохраняем предыдущую серию и начинаем новую
            nums.append(current_num)
            count.append(current_count)
            current_num = x[i]
            current_count = 1
    
    # добавляем последнюю серию
    nums.append(current_num)
    count.append(current_count)
        
print(nums)
print(count)
```

```python
# numpy

x = np.array([2, 2, 2, 3, 3, 3, 5, 6, 6, 7, 2])

# Находим индексы, где элементы меняются
change_indices = np.where(x[1:] != x[:-1])[0] + 1

# Границы серий: начало каждой серии
starts = np.concatenate(([0], change_indices))

# Значения - первые элементы каждой серии
nums = x[starts]

# Длины серий
count = np.diff(np.concatenate((starts, [len(x)])))

print(nums)
print(count)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import scipy.spatial.distance
import time

X = np.random.rand(100, 10).tolist()
Y = np.random.rand(80, 10).tolist()

print(f"Размеры: X {len(X)}x{len(X[0])}, Y {len(Y)}x{len(Y[0])}")
```

```python
# without numpy

start_time = time.time()
n_X = len(X)
n_Y = len(Y)
result_no_numpy = [[0.0] * n_Y for _ in range(n_X)]
    
for i in range(n_X):
    for j in range(n_Y):
        # Вычисляем евклидово расстояние между X[i] и Y[j]
        squared_distance = 0
        for k in range(len(X[i])):
            squared_distance += (X[i][k] - Y[j][k]) ** 2
        result_no_numpy[i][j] = (squared_distance) ** 0.5
time_no_numpy = time.time() - start_time

print(f"Первые 2x2 элемента без numpy:")
for i in range(2):
    print([f"{result_no_numpy[i][j]:.4f}" for j in range(2)])
```

```python
X = np.array(X)
Y = np.array(Y)

start_time = time.time()

# расширяем размерности массивов для попарного сравнения всех элементов
X_expanded = X[:, np.newaxis, :]  # # каждая строка X сравнивается с каждой строкой Y
Y_expanded = Y[np.newaxis, :, :]  # создаем "копии" Y для каждой строки X
    
# вычисляем матрицу расстояний
squared_diff = np.sum((X_expanded - Y_expanded) ** 2, axis=2)
result_numpy = np.sqrt(squared_diff)
time_numpy = time.time() - start_time

print(f"\nПервые 2x2 элемента с numpy:")
print(result_numpy[:2, :2])

start_time = time.time()
result_scipy = scipy.spatial.distance.cdist(X, Y, metric='euclidean')
time_scipy = time.time() - start_time

print(f"\nПервые 2x2 элемента scipy:")
print(result_scipy[:2, :2])
```

```python
print('Время работы')
print(f"Без numpy: {time_no_numpy:.6f} сек")
print(f"С numpy:   {time_numpy:.6f} сек")
print(f"Scipy:     {time_scipy:.6f} сек")
```
